## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

# Q2. Find the top 10 major project themes (using column 'mjtheme_namecode').

```python
# use normalization to create a DataFrame for 'mjtheme_namecode'
mjtheme_df = json_normalize(json_data, 'mjtheme_namecode')
mjtheme_df
```

```python
from collections import defaultdict
```

```python
mjtheme_dict = defaultdict(str)
mjtheme_dict = mjtheme_df[mjtheme_df['name'] != '']
mjtheme_dict = mjtheme_dict.drop_duplicates().set_index('code')
mjtheme_dict
```

```python
mjtheme_count_df = mjtheme_df.groupby('name').count()
```

```python
mjtheme_count_df.columns = ['Number of Projects']
```

```python
mjtheme_count_df.index.name = 'Project Themes'
```

```python
mjtheme_count_df = mjtheme_count_df.sort_values(ascending = False, by ='Number of Projects').head(10)
```

```python
# the top 10 major project themes 
mjtheme_count_df
```

# Q3.  Create a dataframe with the missing names filled in.

```python
mjtheme_df
```

```python
# create an empty dictionary
missing = {}
```

```python
# for loop to match codes with missing names
missing_names = mjtheme_df.values
for code, name in missing_names:
    if name != '':
        missing[code] = name
```

```python
missing
```

```python
# for loop to fill empty cells with correct names
for entry in missing_names:
    if entry[1] == '':
        entry[1] = missing[entry[0]]
```

```python
missing_names
```

```python
# create a Pandas DataFrame with missing names filled in
missing_names_filled = pd.DataFrame(missing_names, columns = ['code', 'name'])
```

```python
# DataFrame with all missing names filled in
missing_names_filled
```
